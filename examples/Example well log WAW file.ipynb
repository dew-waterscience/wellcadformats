## Create WAW object/file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
depths = np.linspace(0, 5, 50)
depths
```

```python
log_data = np.random.normal(size=len(depths))
log_data += 1
log_data *= 120
log_data += 100
log_data
```

```python
fig = plt.figure(figsize=(5, 2))
ax = fig.add_subplot(111)
ax.plot(depths, log_data)
ax.set_xlabel('depth')
ax.set_ylabel('log data')
```

```python
df = pd.DataFrame({"depth": depths, "log_data": log_data})
log = wellcadformats.WAW.from_df(df, units={"log_data": "cps"})
```

```python
log
```

```python
log.df
```

```python
log.to_las(sys.stdout)
```
